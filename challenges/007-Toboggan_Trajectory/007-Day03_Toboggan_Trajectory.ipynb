## Solution - Part 1

> To run this script, you need to save ```input-day-03.txt``` and run it in your local machine.

````python
map = []
start = [0, 0]
number_tree = 0

with open ("input-day-03.txt", "r") as file:
    map = file.read().splitlines()
    
total_rows = len(map)
total_columns = len(map[0])

def goToRight(start):
    start[0] = start[0] + 1
    start[1] = (start[1] + 3) % total_columns
    return start

def setMark(position):
    square_type = map[position[0]][position[1]]
    if square_type == '.':
        temp_row = list(map[position[0]])
        temp_row[position[1]] = "O"        
        map[position[0]] = "".join(temp_row)
        return 0
    else:
        temp_row = list(map[position[0]])
        temp_row[position[1]] = "X"        
        map[position[0]] = "".join(temp_row)
        return 1

countinue_flag = True

while countinue_flag:    
    position = goToRight(start)    
    if position[0] < total_rows:
        number_tree = number_tree + setMark(position)   
    else:
        countinue_flag = False

print("Number of Trees: {}".format(number_tree))
````

## Problem - Part 2

Time to check the rest of the slopes - you need to minimize the probability of a sudden arboreal stop, after all.

Determine the number of trees you would encounter if, for each of the following slopes, you start at the top-left corner and traverse the map all the way to the bottom:

- Right 1, down 1.
- Right 3, down 1. (This is the slope you already checked.)
- Right 5, down 1.
- Right 7, down 1.
- Right 1, down 2.

In the above example, these slopes would find 2, 7, 3, 4, and 2 tree(s) respectively; multiplied together, these produce the answer 336.

What do you get if you multiply together the number of trees encountered on each of the listed slopes?

## Solution - Part 2

> To run this script, you need to save ```input-day-03.txt``` and run it in your local machine.

````python
initial_map = []
map = []
row = 0
column = 0

start = [row, column]
number_tree = 1

slope_list = [[1,1],[1,3],[1,5],[1,7],[2,1]]

with open ("input-day-03.txt", "r") as file:
    initial_map = file.read().splitlines()
    
total_rows = len(initial_map)
total_columns = len(initial_map[0])

def goToRight(start, row, column):
    start[0] = start[0] + row
    start[1] = (start[1] + column) % total_columns
    return start

def setMark(position):
    square_type = map[position[0]][position[1]]
    if square_type == '.':
        temp_row = list(map[position[0]])
        temp_row[position[1]] = "O"        
        map[position[0]] = "".join(temp_row)
        return 0
    else:
        temp_row = list(map[position[0]])
        temp_row[position[1]] = "X"        
        map[position[0]] = "".join(temp_row)
        return 1

for slope in slope_list:
    slope_column = slope[1]
    slope_row = slope[0]
    countinue_flag = True
    temp_number_tree = 0
    map = initial_map.copy()
    start = [0, 0]
    while countinue_flag:    
        position = goToRight(start, slope_row, slope_column)
    
        if position[0] < total_rows:
            temp_number_tree = temp_number_tree + setMark(position)   
        else:
            countinue_flag = False
    number_tree = number_tree * temp_number_tree
    print("Number of Trees for slope[{},{}]: {}".format(slope_row, slope_column, temp_number_tree))
    
print("Total Multiplication: {}".format(number_tree))
````
