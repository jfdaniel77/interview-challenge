## Problem - Part 1
You land at the regional airport in time for your next flight. In fact, it looks like you'll even have time to grab some food: all flights are currently delayed due to issues in luggage processing.

Due to recent aviation regulations, many rules (your puzzle input) are being enforced about bags and their contents; bags must be color-coded and must contain specific quantities of other color-coded bags. Apparently, nobody responsible for these regulations considered how long they would take to enforce!

For example, consider the following rules:

```
light red bags contain 1 bright white bag, 2 muted yellow bags.
dark orange bags contain 3 bright white bags, 4 muted yellow bags.
bright white bags contain 1 shiny gold bag.
muted yellow bags contain 2 shiny gold bags, 9 faded blue bags.
shiny gold bags contain 1 dark olive bag, 2 vibrant plum bags.
dark olive bags contain 3 faded blue bags, 4 dotted black bags.
vibrant plum bags contain 5 faded blue bags, 6 dotted black bags.
faded blue bags contain no other bags.
dotted black bags contain no other bags.
```

These rules specify the required contents for 9 bag types. In this example, every faded blue bag is empty, every vibrant plum bag contains 11 bags (5 faded blue and 6 dotted black), and so on.

You have a shiny gold bag. If you wanted to carry it in at least one other bag, how many different bag colors would be valid for the outermost bag? (In other words: how many colors can, eventually, contain at least one shiny gold bag?)

In the above rules, the following options would be available to you:

```
A bright white bag, which can hold your shiny gold bag directly.
A muted yellow bag, which can hold your shiny gold bag directly, plus some other bags.
A dark orange bag, which can hold bright white and muted yellow bags, either of which could then hold your shiny gold bag.
A light red bag, which can hold bright white and muted yellow bags, either of which could then hold your shiny gold bag.
```

So, in this example, the number of bag colors that can eventually contain at least one shiny gold bag is 4.

How many bag colors can eventually contain at least one shiny gold bag? (The list of rules is quite long; make sure you get all of it.)

## Solution - Part 1

> To run this script, you need to save ```input-day-07.txt``` and run it in your local machine.

````python
import re

# Get data from file
def import_input_data():
    with open ("input-day-07.txt", "r") as file:
        input_data = file.read().splitlines()
    return input_data

# populate data into this format
# {
#   "light red": ["bright white", "muted yellow"]
# }
def populate_data_structure(input_data):
    
    bag_content_dict = {}
    
    for item in input_data:    
        # Find bag type
        bag_type_regex = r"^(\w+ \w+) bags contain (.*)"
        bag_type_content = re.search(bag_type_regex, item)

        bag_type = bag_type_content[1]
        bag_content = bag_type_content[2][:-1]

        if bag_type not in bag_content_dict:
            bag_content_dict[bag_type] = []

        if "no other bags" in bag_content:        
            bag_content_dict[bag_type] = []
        else:
            temp_content = bag_content.split(", ")

            if type(temp_content) is list:            
                for temp_item in temp_content:                
                    bag_content_dict[bag_type].append(build_bag_content(temp_item.strip()))
            else:            
                bag_content_dict[bag_type].append(build_bag_content(temp_content))
                
    return bag_content_dict

# Split between number of bag and bag type
def build_bag_content(bag_item):
    number = bag_item[:2]
    temp_item = bag_item[2:].rsplit(' ', 1)[0]
    return temp_item


input_data = import_input_data()
bag_list = populate_data_structure(input_data)
target_bag = 'shiny gold'
total_count = 0

# Main process to calculate the number of target bag, e.g. shiny gold
def calculate_target_bag(bag):    
    if bag == target_bag:
        return 1
    
    if bag_list.get(bag) == None:
        return 0
    else:
        counts = []
        for item in bag_list[bag]:
            counts.append(calculate_target_bag(item))
        if len(counts) > 0:
            return max(counts)    
    return 0

for bag in bag_list:
    if bag != target_bag:
        count = calculate_target_bag(bag)
        total_count = total_count + count

print("Total {} bags are {}".format(target_bag, total_count))
````

## Problem Part 2

It's getting pretty expensive to fly these days - not because of ticket prices, but because of the ridiculous number of bags you need to buy!

Consider again your shiny gold bag and the rules from the above example:

- faded blue bags contain 0 other bags.
- dotted black bags contain 0 other bags.
- vibrant plum bags contain 11 other bags: 5 faded blue bags and 6 dotted black bags.
- dark olive bags contain 7 other bags: 3 faded blue bags and 4 dotted black bags.

So, a single shiny gold bag must contain 1 dark olive bag (and the 7 bags within it) plus 2 vibrant plum bags (and the 11 bags within each of those): 1 + 1*7 + 2 + 2*11 = 32 bags!

Of course, the actual rules have a small chance of going several levels deeper than this example; be sure to count all of the bags, even if the nesting becomes topologically impractical!

Here's another example:

```
shiny gold bags contain 2 dark red bags.
dark red bags contain 2 dark orange bags.
dark orange bags contain 2 dark yellow bags.
dark yellow bags contain 2 dark green bags.
dark green bags contain 2 dark blue bags.
dark blue bags contain 2 dark violet bags.
dark violet bags contain no other bags.
```

In this example, a single shiny gold bag must contain 126 other bags.

How many individual bags are required inside your single shiny gold bag?

## Solution - Part 2

> To run this script, you need to save ```input-day-07.txt``` and run it in your local machine.

````python
import re

# Get data from file
def import_input_data():
    with open ("input-day-07.txt", "r") as file:
        input_data = file.read().splitlines()
    return input_data

# populate data into this format
# {
#   "light red": ["bright white": "2", "muted yellow": "1"]
# }
def populate_data_structure(input_data):
    
    bag_content_dict = {}
    
    for item in input_data:    
        # Find bag type
        bag_type_regex = r"^(\w+ \w+) bags contain (.*)"
        bag_type_content = re.search(bag_type_regex, item)

        bag_type = bag_type_content[1]
        bag_content = bag_type_content[2][:-1]

        if bag_type not in bag_content_dict:
            bag_content_dict[bag_type] = {}

        if "no other bags" in bag_content:        
            bag_content_dict[bag_type] = {}
        else:
            temp_content = bag_content.split(", ")

            if type(temp_content) is list:    
                temp = {}
                for temp_item in temp_content:                
                    bag_content = build_bag_content(temp_item.strip())
                    temp[bag_content[0]] = bag_content[1]
                    
                bag_content_dict[bag_type] = temp
            else:            
                bag_content = build_bag_content(temp_item.strip())
                bag_content_dict[bag_type] = {
                        bag_content[0]: bag_content[1]
                }
                
    return bag_content_dict

# Split between number of bag and bag type
def build_bag_content(bag_item):
    number = bag_item[:2]
    temp_item = bag_item[2:].rsplit(' ', 1)[0]
    return (temp_item, number)


input_data = import_input_data()
bag_list = populate_data_structure(input_data)

target_bag = "shiny gold"
bags_contains = {}

for k, v in bag_list.items():
    bags_contains[k] = []
    
    for kk, vv in v.items():
        bags_contains[k]+=[kk]*int(vv)

# Main process to calculate the number target bag can hold, e.g. shiny gold
def count_bags(current_bag):
    if current_bag==" " or bags_contains.get(current_bag) is None:
        return 0
    count = len(bags_contains[current_bag])
    bag_container = []
    for k in bags_contains[current_bag]:
        bag_container.append(count_bags(k))    
    return sum(bag_container) + count

print("{} bag can hold {} bags".format(target_bag, count_bags('shiny gold')))
````
