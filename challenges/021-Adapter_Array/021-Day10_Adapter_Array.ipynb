## Problem - Part 1
Patched into the aircraft's data port, you discover weather forecasts of a massive tropical storm. Before you can figure out whether it will impact your vacation plans, however, your device suddenly turns off!

Its battery is dead.

You'll need to plug it in. There's only one problem: the charging outlet near your seat produces the wrong number of jolts. Always prepared, you make a list of all of the joltage adapters in your bag.

Each of your joltage adapters is rated for a specific output joltage (your puzzle input). Any given adapter can take an input 1, 2, or 3 jolts lower than its rating and still produce its rated output joltage.

In addition, your device has a built-in joltage adapter rated for 3 jolts higher than the highest-rated adapter in your bag. (If your adapter list were 3, 9, and 6, your device's built-in adapter would be rated for 12 jolts.)

Treat the charging outlet near your seat as having an effective joltage rating of 0.

Since you have some time to kill, you might as well test all of your adapters. Wouldn't want to get to your resort and realize you can't even charge your device!

If you use every adapter in your bag at once, what is the distribution of joltage differences between the charging outlet, the adapters, and your device?

## Solution - Part 1

> To run this script, you need to save ```input-day-10.txt``` and run it in your local machine.

````python
def import_input_data():
    with open ("input-day-10.txt", "r") as file:
        input_data = file.read().splitlines()
    return input_data

def convert_to_int(input_data):
    input_data = [int(item) for item in input_data]
    return input_data

def add_highest_rated(input_data):
    max_value = max(input_data)
    input_data.append(max_value + 3)
    return input_data

def process_calculation(input_data):
    cursor = 0
    has_next = True
    counter_jolt = {
        "1": 0,
        "2": 0,
        "3": 0
    }
    
    while (has_next):
        for idx in range(1, 4):
            jolt = cursor + idx
            if jolt in input_data:                
                cursor = jolt
                counter_jolt[str(idx)] = counter_jolt[str(idx)] + 1
                break
                
        if cursor >= max(input_data):
            has_next = False
    
    return counter_jolt
    
    
    
# Get data from file
input_data = import_input_data()

# Convert each data to int
input_data = convert_to_int(input_data)

# Add highest-rated adapter
input_data = add_highest_rated(input_data)

# Count 1-jolt and 3-jolt
counter_jolt = process_calculation(input_data)
result = counter_jolt["1"] * counter_jolt["3"]

print("The number of 1-jolt differences multiplied by the number of 3-jolt differences: {}".format(result))
````

## Solution - Part 2

> To run this script, you need to save ```input-day-10.txt``` and run it in your local machine.

````python
class Node:
    
    def __init__(self, name):
        self.outgoing = []
        self.incoming = []
        self.name = name
        
    def get_outgoing(self):
        return self.outgoing
    
    def get_incoming(self):
        return self.incoming
    
    def get_name(self):
        return self.name
    
    def add_edge(self, edge):
        if (edge.get_from() == self):
            self.outgoing.append(edge)
        
        if (edge.get_to() == self):
            self.incoming.append(edge)   
            
    def __repr__(self):
        value = "Node {} has outgoing {}".format(self.name, self.outgoing)
        return value
            
class Edge:
    
    def __init__(self, from_node, to_node):
        self.from_node = from_node
        self.to_node = to_node

    def get_from(self):
        return self.from_node
    
    def get_to(self):
        return self.to_node
    
    def __repr__(self):
        value = "From {} and to {}".format(self.from_node.get_name(), self.to_node.get_name())
        return value
        
class Graph:
    
    def __init__(self):
        self.nodes = []
        self.edges = []
        self.length = 0
        
    def get_nodes(self):
        return self.nodes
    
    def get_edges(self):
        return self.edges
    
    def add_node(self, node):
        self.nodes.append(node)
        self.length += 1
    
    def get_length(self):
        return self.length
        
    def add_edge(self, edge):
        self.edges.append(edge)
        edge.get_from().add_edge(edge)
        edge.get_to().add_edge(edge)
    
    def __repr__(self):
        for node in self.nodes:
            value = "Node {} has {}".format(node.get_name(), node.get_outgoing())
            print(value)
        
def count_paths(graph, source, target):
    visiteds = []
    memoize = {}
    
    for node in graph.get_nodes():   
        memoize[node.get_name()] = -1
    
    memoize[target.get_name()] = 1
    count = BFS(source, visiteds, memoize)    
    return count

def BFS(current, visiteds, memoize):
    if memoize[current.get_name()] != -1:
        return memoize[current.get_name()]
    else:        
        count = 0        
        visiteds.append(current)
        for edge in current.get_outgoing():
            if (edge.get_to not in visiteds):
                count = count + BFS(edge.get_to(), visiteds, memoize)
                
        visiteds.remove(current)
        memoize[current.get_name()] = count
        return count
    
def import_input_data():
    with open ("input-day-10.txt", "r") as file:
        input_data = file.read().splitlines()
    return input_data

def convert_to_int(input_data):
    input_data = [int(item) for item in input_data]
    return input_data

def add_highest_rated(input_data):
    max_value = max(input_data)
    input_data.append(max_value + 3)
    return input_data

def add_initial(input_data):
    input_data.append(0)
    return input_data

input_data = import_input_data()
input_data

# Convert each data to int
input_data = convert_to_int(input_data)

# Add highest-rated adapter
input_data = add_highest_rated(input_data)

# Add initial value
input_data = add_initial(input_data)

# Initialize Graph
graph = Graph()
node = Node("0")

graph.add_node(node)
created_node = {
    "0": node
}

cursor = 0

for i in range(0, max(input_data)+1):
    
    if str(i) in created_node:
        branch_1 = i + 1
        if branch_1 in input_data:
            node = None
            if str(branch_1) in created_node:
                node = created_node[str(branch_1)]
            else:
                node = Node(str(branch_1))
            previous_node = created_node[str(cursor)]
            edge = Edge(previous_node, node);
            graph.add_edge(edge)
            
            if str(branch_1) not in created_node:    
                graph.add_node(node)   
                created_node[str(branch_1)] = node
            
        branch_2 = i + 2
        
        if branch_2 in input_data:
            node = None
            if str(branch_2) in created_node:
                node = created_node[str(branch_2)]
            else:
                node = Node(str(branch_2))
                
            previous_node = created_node[str(cursor)]
            edge = Edge(previous_node, node);
            graph.add_edge(edge)
            if str(branch_2) not in created_node:    
                graph.add_node(node)   
                created_node[str(branch_2)] = node
            
        branch_3 = i + 3
        
        if branch_3 in input_data:
            node = None
            if str(branch_3) in created_node:
                node = created_node[str(branch_3)]
            else:
                node = Node(str(branch_3))
                
            previous_node = created_node[str(cursor)]
            edge = Edge(previous_node, node);
            graph.add_edge(edge)
            if str(branch_3) not in created_node:  
                graph.add_node(node)   
                created_node[str(branch_3)] = node
    cursor += 1

print("Total nodes = {}".format(graph.get_length()))
print("Total distict ways: {}".format(count_paths(graph, created_node["0"], created_node[str(max(input_data))])))
````
